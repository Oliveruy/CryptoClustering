```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

**Prepare the data**

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(scaled_data, columns=df_market_data.columns)

# Copy the crypto names from the index of the original data
df_scaled.index = df_market_data.index

# Display the data
df_scaled.head(5)
```

**Find the Best Value for k Using the Original Data.**

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Display the list of k-values
print(k_values)
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    
    # 2. Fit the model to the data using `df_scaled`
    kmeans_model.fit(df_scaled)
    
    # 3. Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k_values': k_values,
    'inertia_values': inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values
elbow_curve_original = df_elbow.hvplot.line(x='k_values', y='inertia_values', xlabel='Number of Clusters (k)', ylabel='Inertia', 
                     title='Elbow Curve', width=400, height=300, grid=True, line_color='steelblue', line_width=2)

elbow_curve_original
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for 'k'is 4 since it shows a clear elbow point.

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
clusters = kmeans_model.predict(df_scaled)

# Print the resulting array of cluster values
print(clusters)
```

```python
# Create a copy of the DataFrame
df_clustered = df_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_clustered['predicted_cluster'] = clusters

# Display sample data
print(df_clustered.head())
```

```python
# Create the scatter plot
scatter_plot_1 = df_clustered.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="predicted_cluster",
    hover_cols=["coin_id"],
    width=400,
    height=300
)

# Show the scatter plot
scatter_plot_1
```

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance with n_components=3
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with fit_transform to reduce to three principal components
principal_components = pca_model.fit_transform(df_scaled)

# Convert the principal components array to a DataFrame
df_principal_components = pd.DataFrame(
    data=principal_components,
    columns=['PC1', 'PC2', 'PC3'],  # Specify column names for the principal components
    index=df_scaled.index  # Use the same index as the original DataFrame
)

# View the first five rows of the DataFrame
df_principal_components.head()
```

```python
# Retrieve the explained variance for each principal component
explained_variance = pca_model.explained_variance_ratio_

# Convert the explained variance to a DataFrame for better readability
df_explained_variance = pd.DataFrame(
    data=explained_variance,
    columns=['Explained Variance'],
    index=['PC1', 'PC2', 'PC3']  # Specify the principal component names as index
)

# Display the DataFrame
print(df_explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** Using the below code, the total explained variance is 0.895031657030984.

```python
# Sum up the explained variance for all three principal components
total_explained_variance = explained_variance.sum()

# Print the total explained variance
print("Total Explained Variance:", total_explained_variance)
```

```python
# Creating a DataFrame with the PCA data
df_pca_data = pd.DataFrame(
    data=principal_components,  # PCA data
    columns=['PC1', 'PC2', 'PC3'],  # Principal component column names
    index=df_scaled.index  # Use the same index as the original DataFrame
)

# Copy the crypto names from the original data
df_pca_data['coin_name'] = df_scaled.index

# Set the coinid column as the index
df_pca_data.set_index('coin_name', inplace=True)

# Display sample data
df_pca_data.head()
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values = range(1, 12)
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # Create a KMeans model with the current value of k
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model to the PCA data and compute inertia
    kmeans_model.fit(df_pca_data)
    
    # Append the model's inertia to the inertia list
    inertia_values.append(kmeans_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {
    'k_values': k_values,
    'inertia_values': inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
pca_df_elbow_curve = pd.DataFrame(pca_elbow_data)
```

```python
# Plot a line chart with all the inertia values
elbow_curve_pca = pca_df_elbow_curve.hvplot.line(
    x='k_values',
    y='inertia_values',
    title='Elbow Curve',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    grid=True,
    width=400,
    height=300
)

# Show the plot
elbow_curve_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** best `k` value when using PCA data is 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No , it is the same.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
kmeans_model.fit(df_pca_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
clusters = kmeans_model.predict(df_pca_data)

# Print the resulting array of cluster values
print(clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
df_clustered = df_pca_data.copy()

# Add a new column to the DataFrame with the predicted clusters
df_clustered['predicted_cluster'] = clusters

# Display sample data
df_clustered.head()
```

```python
# Create a scatter plot using hvPlot
scatter_plot_2 = df_clustered.hvplot.scatter(
    x='PC1',
    y='PC2',
    c='predicted_cluster',
    colormap='viridis',
    hover_cols=['coin_name'], 
    width=400,
    height=300
)

# Show the scatter plot
scatter_plot_2
```

```python
# Composite plot to contrast the Elbow curves
# Combine the two elbow plots
elbow_composite_plot = elbow_curve_original + elbow_curve_pca

# Display the composite plot
elbow_composite_plot
```

```python
# Composite plot to contrast the clusters
# Combine the two elbow plots
scatter_composite_plot = scatter_plot_1 + scatter_plot_2

# Display the composite plot
scatter_composite_plot
```
